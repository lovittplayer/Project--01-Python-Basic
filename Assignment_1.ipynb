```python
L = [11,12,13,14]
print("Original List -->",L)
L1 = [50,60]
L.extend(L1) ## use of extend instead of appened so that there is no issue while sorting the List
print("After adding 50 and 60 -->",L)
del L[0]
del L[1]
print("After deleting 11 and 13 -->",L)

L2 = sorted(L)
print("Ascending Order -->",L2)

L.sort(reverse=True)
print("Descending Order -->",L)

if 13 in L:
  print("13-->Present")
else:
  print("13-->Not Present")


print("Number of elements in List --> ",len(L))

OddSum = 0

for k in L:
  if(k%2!=0):
   OddSum += k


print("The sum of odd numbers of List is -->",OddSum)




EveSum = 0
for k in L:
  if(k%2==0):
   EveSum += k


print("The sum of Even numbers of List is -->\t",EveSum)


PrimeSum = 0
for k in L:
    if k < 2:
        continue  # Not prime
    is_prime = True
    for i in range(2, int(k**0.5) + 1):
        if k % i == 0:
            is_prime = False
            break
    if is_prime:
        PrimeSum += k


print("The sum of Prime numbers of List is -->\t", PrimeSum)

L.clear()
print("Elements of L-->",L)
del L
print("List is deleted")




```
